```python
import pandas as pd   #for reading csv data
import matplotlib.pyplot as plt
```

```python
#load the data from the csv file to the jupyter notebook
data=pd.read_csv('/Users/delta/python/unemployment_analysis.csv')
```

```python
#clear the unwanted data
data.drop(columns=["Country Code"], inplace=True)
data.drop(columns=["1991","1992","1993","1994","1995","1996","1997","1998","1999","2000","2001","2002","2003","2004","2005","2006","2007"],inplace=True)

#display the data to check the drop
data
```

```python
#modifying data for better understanding
data = data.set_index("Country Name").transpose()
data.index.names = ["Year"]
data.head()
```

```python
#Creating a dataFrame with few selected countries
df=data.loc[:,["Afghanistan","Australia","United States","Pakistan","India","Vietnam"]]
print(df)
```

```python
#plot the graph for unemployment rate from 2008 to 2021 for the selected coutries
df["0":"5"].plot()
plt.show()
```
