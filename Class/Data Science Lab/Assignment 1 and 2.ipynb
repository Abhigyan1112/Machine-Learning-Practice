```python
df_num=df.copy()
for col in df_num.columns:
    if df_num[col].dtype==object:
        df_num=df_num.drop(col,axis=1)
```

```python
from sklearn.impute import SimpleImputer

imputer=SimpleImputer(strategy="mean")
imputer.fit(df_num)
```

```python
imputer.statistics_
```

```python
X=imputer.transform(df_num)
df_transformed=pd.DataFrame(X,columns=df_num.columns,index=df_num.index)
```

```python
df_transformed.isnull().sum()
```

```python
df_cat=df.copy()
for col in df_cat.columns:
    if df_cat[col].dtype!=object:
        df_cat=df_cat.drop(col,axis=1)
```

```python
imputer=SimpleImputer(strategy="most_frequent")
imputer.fit(df_cat)
imputer.statistics_
```

```python
X=imputer.transform(df_cat)
df_cat_transformed=pd.DataFrame(X,columns=df_cat.columns,index=df_cat.index)
```

```python
df_cat_transformed.isnull().sum()
```

```python
df_trans=pd.concat([df_transformed,df_cat_transformed],axis=1)
df_trans
```

```python
def Euc_dist(point1,point2):
    return np.sqrt(np.sum((np.array(point1)-np.array(point2))**2))

def Man_dist(point1,point2):
    return sum(abs(a-b) for a,b in zip(point1,point2))
```
