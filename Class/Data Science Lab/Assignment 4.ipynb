```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Online Sales Data.csv")
```

```python
df.describe()
```

```python
numerical=['Units Sold','Unit Price','Total Revenue']
for column in numerical:
    plt.figure(figsize=(6,4))
    sns.histplot(df[column])
    plt.show()
```

```python
for column in numerical:
    plt.figure(figsize=(6,4))
    sns.boxplot(df[column])
    plt.show()
```

```python
for column in numerical:
    plt.figure(figsize=(6,4))
    sns.kdeplot(df[column])
    plt.show()
```

```python
np.random.seed(0)  
dataset = np.random.uniform(0, 1, 1000)

sample_sizes = [5, 20, 100, 1000]
means = {}

for size in sample_sizes:
    sample_means = []
    for _ in range(1000): 
        sample = np.random.choice(dataset, size, replace=True)
        sample_means.append(np.mean(sample))
    means[size] = sample_means

fig, axes = plt.subplots(1, len(sample_sizes), figsize=(15, 4),)
plt.suptitle('Distribution of Sample Means')

for i, size in enumerate(sample_sizes):
    axes[i].hist(means[size], bins=30, alpha=0.7, color='blue')
    axes[i].set_title(f'Sample Size = {size}')
    axes[i].set_xlabel('Mean')
    axes[i].set_ylabel('Frequency')

plt.tight_layout()
plt.subplots_adjust(top=0.85)
plt.show()
```
