```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv("Online Sales Data.csv")
```

```python
df.tail(-60)
```

```python
df.describe()
```

```python
plt.figure(figsize=(6,4))
plt.bar(df['Product Category'],df['Total Revenue'],color="green",width=0.5)
plt.title("Sales")
plt.xlabel("Product Categories")
plt.ylabel("Total Revenue")
plt.xticks(df['Product Category'], rotation=30)
plt.show()
```

```python
df=pd.read_csv("goldstock v1.csv")
df.head(2)
```

```python
plt.figure(figsize=(6, 4))
plt.plot(df['Date'], df['High'], marker='o', color='b')
plt.plot(df['Date'], df['Low'],color='r')
plt.grid(True)
plt.title('High Over Time')
plt.xlabel('Date')
plt.ylabel('High')
plt.xticks(rotation=90)  
plt.show()
```

```python
df=pd.read_csv("Housing.csv")
```

```python
plt.figure(figsize=(6,4))
plt.scatter(df['area'],df['price'])
plt.title("House Size vs Price")
plt.show()
```

```python
df=pd.read_csv("heart_disease.csv")
df.head(2)
```
