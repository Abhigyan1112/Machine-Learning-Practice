### Outliers

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='area',data=df)
plt.show()
```

```python
print(df['area'].quantile(0.95))
```

```python
df['area']=np.where(df['area']>9000,9000,df['area'])
```

### Missing Values

```python
df.isna().sum()
```

```python
for column in df.columns:
    if(column=='area'):
        df[column]=df[column].fillna(df[column].mean())
    else:
        df[column]=df[column].fillna(df[column].mode()[0])
```

### Handling Categorical Data

```python
from sklearn import preprocessing
label_encoder=preprocessing.LabelEncoder()
ordinal_encoder=preprocessing.OrdinalEncoder()
def label_encoding(column):
    df[column]=label_encoder.fit_transform(df[column])

def ordinal_encoding(column):
    df[column]=ordinal_encoder.fit_transform(df[[column]])
```

```python
label_encoding('mainroad')
label_encoding('guestroom')
label_encoding('basement')
label_encoding('hotwaterheating')
label_encoding('airconditioning')
ordinal_encoding('prefarea')
ordinal_encoding('furnishingstatus')
```

### Standardization

```python
def standardization(column):
    df[column]=((df[column]-df[column].mean())/df[column].std())
```

```python
for column in df.columns:
    if df[column].max()>2:
        standardization(column)
```

### Training the model

```python
y=df['price']
X=df.drop('price',axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression().fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
print("RMSE:",np.sqrt(mean_squared_error(y_test,y_pred)))
```

```python
plt.figure(figsize=(6,4))
plt.scatter(y_test, y_pred)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color="red", linestyle="--")  # 45-degree line
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Actual vs. Predicted")
plt.show()
```
