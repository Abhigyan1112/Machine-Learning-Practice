```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

Num_of_hours_studied=np.array([2,3.5,5,6.5,8,9.5]).reshape(-1,1)
Score=([50,60,75,80,90,95])

model=LinearRegression()
model.fit(Num_of_hours_studied,Score)

your_hours_studied=np.array([2,3]).reshape(-1,1)
predicted_Score=model.predict(your_hours_studied)

print(predicted_Score)
print("Intercept=",model.intercept_)
print("Slope=",model.coef_[0])

plt.scatter(Num_of_hours_studied,Score,color="pink",label='Data')
plt.plot(Num_of_hours_studied,model.predict(Num_of_hours_studied),color="purple",label="Regression Line")
plt.scatter(your_hours_studied,predicted_Score,color="green",label="Predictions")
plt.xlabel('Number Of Hours Studied')
plt.ylabel("Score")
plt.show()

```
